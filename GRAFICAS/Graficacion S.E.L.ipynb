```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
A =np.array([[2,4,6],[3,8,5],[-1,1,2]])
b =np.array([22,27,2])
ans=np.linalg.solve(A,b)
print("RESULTADO: {}".format(ans))
```

```python
x,y=np.linspace(0,10,10),np.linspace(0,10,10)
#Creamos una cuadricula
X,Y=np.meshgrid(x,y)
#DESPEJAMOS LAS ECUACIONES PARA "Z"
Z1=(22-2*X-4*Y)/6
Z2=(-12 + 4*X - 4*Y) / 8
Z3 =(2 + X - Y) / 2
```

```python
#GRAFICA
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z1,alpha=0.5,cmap=cm.flag,rstride=100,cstride=100)
ax.plot_surface(X,Y,Z2,alpha=0.5,cmap=cm.Paired,rstride=100,cstride=100)
ax.plot_surface(X,Y,Z3,alpha=0.5,cmap=cm.Pastel1,rstride=100,cstride=100)
#INTERSECCION
ax.plot((ans[0],),(ans[1],),(ans[2],),lw=2,c='k',marker='o',markersize=7,markeredgecolor='g',markerfacecolor='white')
#Nombres de los ejes.
ax.set_xlabel("X");ax.set_ylabel("Y");ax.set_zlabel("Z");
plt.show()
```
