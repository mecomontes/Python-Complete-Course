# ¿Cómo hacer varias gráficas en una sola ventana?

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Declaramos el objeto de la figura
fig = plt.figure(figsize=(15,15))
#Definimos el espacio suficiente de las graficas.
fig.tight_layout()
#Definimos una lista de colores para hacer la difernecia cuando se defina el objeto color
colores =  ['blue','green','red','cyan','magenta','yellow','black','white']
```

```python
#Inicia Ciclo For.
for i in range(1,7):
    x = np.arange(0,16,2)#Creamos un rango de numeros
    #Creamos ecuacion aleatoria.
    y = pow(x,np.random.randint(0,6))+np.random.randint(-5,10)*x
    #Creamos objeto ax 
    ax = plt.subplot(2,3,i)#De dos a 3 columnas y que se acomode i veces
    ax.plot(x,y,color = colores[i])
    ax.set_xlabel('x')
    ax.set_xlabel('y')
    ax.set_title('Grafica' + str(i))#Imprime grafica 1,2,3.n
    
```
