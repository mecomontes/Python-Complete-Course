# ¿Cómo acomodar las gráficas y cambiar el formato de los ejes? 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker #Permite cambiar los valores de los ejes
```

```python
x = np.arange(-10,10.5,.5)
#Generamos ecuacion cubica para poder visualizar mejor la grafica
y = pow(x,3)+pow(x,2)+.5*x
z = pow(x,3)+pow(x,2)+x
#Creamos objeto para visualizar las graficas
fig = plt.figure(figsize=(15,15))
fig.tight_layout()#Ajusta bien el valor para que no se empalme
#Creamos graficas
ax1 = fig.add_subplot(1,2,2)#un renglon 2 columnas y donde aparece el valor
ax2 = fig.add_subplot(1,2,1)
ax1.plot(x,y)
ax2.plot(x,z)
#Agregamos etiquetas de la grafica
ax1.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.set_title('Grafica 1')
ax2.set_title('Grafica 2')
#Modificamos escala de el eje x
escala = 10
ticks_x = ticker.FuncFormatter(lambda x,pos:'{:.2f}'.format(x/10))
ax1.xaxis.set_major_formatter(ticks_x)


ticks_y = ticker.FuncFormatter(lambda y,pos:'{:.2f}'.format(y/10))
ax1.xaxis.set_major_formatter(ticks_y)
#mostramos grafica
plt.show()
```
