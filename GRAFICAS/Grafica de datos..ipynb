# Grafica de barras horizontales(groupby & plot)

```python
df=pd.DataFrame(data)
df.groupby('genres')['gross'].sum().plot(kind='barh',legend='Reverse')
plt.xlabel('Ganancias')
```

# Grafica de Pay (groupby & plot)

```python
df.movie_facebook_likes.groupby(df.genres).mean().plot(kind='pie',cmap='Set2')
plt.axis('equal')
plt.ylabel('')
plt.title("Promedio de likes de Facebook")
```

# Seaborn.

# Grafica de dispercion(Seaborn).

```python
#FORMA 1
sns.lmplot(x='num',y='gross',data=df,fit_reg=False,hue='num',legend=False,palette='Set2') #Este metodo hace correlaciones de tipo lineal
```

# Gráfica Pairplot utilizando Seaborn.

```python
#FORMA2
new=df[['genres','movie_facebook_likes','actor_2_facebook_likes']]
sns.set(style='ticks',color_codes=True)
g=sns.pairplot(new,hue='genres',palette='Spectral')
```

# Gráficas Lmplot (Correlación lineal) con Seaborn.

```python
g=sns.lmplot(x='duration',y='movie_facebook_likes',data=df,palette='Set1')
```

```python
g=sns.lmplot(x='gross',y='movie_facebook_likes',data=df,hue='num',col='title_year',aspect=.4,x_jitter=0.1)
```

```python
g=sns.lmplot(x='gross',y='movie_facebook_likes',hue='content_rating',data=df,palette='Set1')
```
