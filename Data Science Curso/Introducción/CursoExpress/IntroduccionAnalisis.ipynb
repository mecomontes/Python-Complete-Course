```python
# importamos la libreria a utilizar.
import pandas as pd
```

<h2> Leer datos </h2>
<p>
Usamos la función <code> pandas.read_csv () </code> para leer el archivo csv. En el paréntesis, colocamos la ruta del archivo junto con una comilla, para que los pandas lean el archivo en un marco de datos desde esa dirección. La ruta del archivo puede ser una URL o su dirección de archivo local. <br>
Como los datos no incluyen encabezados, podemos agregar un argumento <code> headers = None </code> dentro del método <code> read_csv () </code>, para que los pandas no establezcan automáticamente la primera fila como encabezado . <br>
También puede asignar el conjunto de datos a cualquier variable que cree.
</p>

```python
other_path = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
data = pd.read_csv(other_path, header=None)
```

```python
#Mostramos las primeras 5 columas de data
print("Las primeras 5 columnas de data") 
data.head(5)
```

<h3> Agregar encabezados </h3>

En primer lugar, creamos una lista de "encabezados" que incluyen todos los nombres de columna en orden.
Luego, usamos <code> dataframe.columns = headers </code> para reemplazar los encabezados por la lista que creamos.
</p>

```python
encabezados = ["simbolización", "pérdidas normalizadas", "marca", "tipo de combustible", "aspiración", "número de puertas", "estilo de carrocería",
         "ruedas motrices", "ubicación del motor", "distancia entre ejes", "longitud", "ancho", "altura", "peso en vacío", "tipo de motor",
         "número de cilindros", "tamaño del motor", "sistema de combustible", "diámetro interior", "carrera", "relación de compresión", "potencia",
         "pico-rpm", "ciudad-mpg", "carretera-mpg", "precio"]
print("Encabezados\n", encabezados)
```

Reemplazamos los encabezados y volvemos a verificar nuestro marco de datos

```python
data.columns = encabezados
data.head(10)
```

 podemos soltar los valores faltantes a lo largo de la columna "precio" de la siguiente manera

```python
data.dropna(subset=["precio"], axis=0)
```

<h2> Guardar conjunto de datos </h2>
<p>
En consecuencia, Pandas nos permite guardar el conjunto de datos en csv utilizando el método <code> dataframe.to_csv () </code>, puede agregar la ruta del archivo y el nombre junto con comillas entre paréntesis.
</p>
<p>
    Por ejemplo, si guarda el marco de datos <b> df </b> como <b> automobile.csv </b> en su máquina local, puede usar la sintaxis a continuación:
</p>

```python
data.to_csv("automobile.csv", index=False)
```

También podemos leer y guardar otros formatos de archivo, podemos usar funciones similares a pd.read_csv () y df.to_csv () para otros formatos de datos, las funciones se enumeran en la siguiente tabla:

<h2>Leer / guardar otros formatos de datos</h2>

<h2> Tipos de datos </h2>
<p>
Los datos tienen una variedad de tipos. <br>
Los principales tipos almacenados en los marcos de datos de Pandas son <b> objeto </b>, <b> flotante </b>, <b> int </b>, <b> bool </b> y <b> datetime64 </ b>. Para conocer mejor cada atributo, siempre es bueno para nosotros conocer el tipo de datos de cada columna. En pandas:
</p>

```python
data.dtypes
```

<h2> Describe </h2>
Si nos gustaría obtener un resumen estadístico de cada columna, como conteo, valor medio de columna, desviación estándar de columna, etc. Utilizamos el método de descripción:

```python
data.describe(include = "all")
```

<p>
Ahora, proporciona el resumen estadístico de todas las columnas, incluidos los atributos de tipo objeto. <br>
Ahora podemos ver cuántos valores únicos, que es el valor superior y la frecuencia del valor superior en las columnas de tipo objeto. <br>
Algunos valores en la tabla anterior se muestran como "NaN", esto se debe a que esos números no están disponibles con respecto a un tipo de columna en particular. <br>
</p>
