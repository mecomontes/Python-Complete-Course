# Generar Numeros Aleatorios

Un número aleatorio es una cantidad generada por un proceso que simula el azar, como cambiar canciones en modo “random” o “shuffle”. Estos procesos se calculan con algoritmos complejos como el Mersenne Twister o el Fisher-Yates. Los números aleatorios pueden echarnos una mano en muchas situaciones en las que queremos probar métodos nuevos.

```python
import numpy as np
```

```python
#Con esta fucnion generamos un numero aleatorio entero
np.random.randint(1,32)
```

```python
np.random.random()
```

```python
##Funcion que genera una lista de n numeros aleatorios dentro del intervalo [a,b]
def randint_list(n,a,b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x    
```

```python
randint_list(5,3,303)
```

```python
#Con la funcion randrange podemos generar numeros aleatorios si hacer una funcion
for i in range(10):
     print(random.randrange(0,70,3))
```
