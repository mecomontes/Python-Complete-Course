# Agregacion de datos


La agregación de datos es un tipo de proceso de minería de datos e información en el que los datos son buscados, recopilados y presentados en un formato resumido basado en informes para lograr objetivos o procesos comerciales específicos y / o realizar análisis humanos.

```python
import numpy as np
import pandas as pd
```

```python
#Creo 2 categorias.
gender = ["Male","Female"]
income = ["Poor","Midle","Rich"]
```

```python
n = 500
gender_data = []
income_data = []

#Creamos los 500 datos aleatorios con un bucle
for i in range(0,n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
income_data[1:5]
```

```python
#Generamos datos nuevos
height = 160 + 30*np.random.randn(n)
weight = 65 + 25*np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
#Incorporamos todos los valores al data frame
data = pd.DataFrame(
  {
    "Gender": gender_data,
    "Economic Status": income_data,
    "Height": height,
    "Weight": weight,
    "Age": age,
    "Income": income
      
  }
)
```

```python
data.head()
```

# Agrupacion de datos.

Una agrupacion de datos es una de las formas más sencillas de presentarlos: organiza los valores en orden ascendente o descendente.

La agrupacion de datos ofrece varias ventajas con respecto a los datos sin procesar:

1.- Podemos identificar los valores de mayor a menor rápidamente.

2.- Es fácil dividir los datos en secciones.

3.- Podemos ver si algunos valores aparecen más de una vez en el arreglo

4.- Podemos observar la distancia entre valores sucesivos de los datos.

```python
grouped_gender= data.groupby("Gender")
```

```python
#Podemos observar a que grupos pertenecen las filas 
grouped_gender.groups
```

```python
#Iteramos sobre cada una de esas opciones para montar un recorrido sobre cada una de las grupaciones posbiles para hacer operaciones sparadas para cada una de los grupos
for names, groups in grouped_gender:
    print(names)
    print(groups)
    
#Podemos observar que primero se imprime toda la categoria femina y despues la categoria msaculina    
```

```python
#Solo tomamos el caso de las mujeres con la funcion get_group
grouped_gender.get_group("Female")
```

```python
#Agrupamos los datos de 2 categorias
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

# Filtrado de datos

un filtro de datos  filtra datos del dataset
con el propósito de devolver la información que responda exactamente a lo que se necesita. Al analizar un informe, es importante conocer las condiciones del filtro que se han aplicado al informe. Esto proporciona un mejor conocimiento de los datos que se han excluido de los resultados.

```python
double_group["Income"].filter(lambda x: x.sum()>30000)
```

# Transformacion de variables.

```python
zscore = lambda x : (x - x.mean())/x.std()
```

```python

z_group = double_group.transform(zscore)
```
