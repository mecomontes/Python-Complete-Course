# Introduccion Basica a la regresion Lineal.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importamos las libreria que vamos a utilizar
from sklearn.datasets import load_boston
```

```python
data = load_boston()
#Vemos como esta montada la informacion de nuestro dataset
print(data.DESCR)
```

Formula de minimizar el error cuadratico medio(MCD)

$\beta =({x}^{T}X)^{-1}X^{T}Y$

```python
#Extraemos las columnas que nos interesan, en este caso el numero de habitaciones.
X = np.array(data.data[:,5])
Y = np.array(data.target)
#Graficamos.
plt.scatter(X,Y,alpha=0.5,facecolor='cyan',edgecolors="black")
plt.xlabel("Numero Habitaciones")
plt.ylabel("Costo Vivienda")



#Añadimos columna 1 para los terminos independientes.
X = np.array([np.ones(506),X]).T

# Escribimos la formula MCD

B = np.linalg.inv(X.T @ X) @ X.T @ Y
plt.plot([4,9],[B[0]+B[1]*4,B[0]+B[1]*9],color="red")
plt.show()
#Podemos ver la linea de regresion por metodo de cuadrados ordinarios.
```
