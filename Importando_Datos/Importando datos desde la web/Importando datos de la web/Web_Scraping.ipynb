Ahora  tenemos el HTML de nuestra pagina de interés  pero ,en general HTML es un revoltijo de datos estructurados y no estrucuturados , recoredemos que los datos estructurados son datos que tienen un modelo de datos predefinido o que se organiza de manera definida.Por otro lado los datos no estructurados son datos que no poseen ninguna de esas propiedades.

HTML es interesante porque ,aunque gran parte es texto no estructurado, contiene etiquetas dónde por ejemplo , se pueden encontrar encabezados e hipervínculos.

En general , para convertir HTML que hemos extraido de la web , deberemos analizarlos y extraer datos estrucutrados de ellos.


A continuación veremos un breve resumen de como hacerlo,utilizando el paquete de Python BeautifulSoup.

En wl desarrollo web , el términuo 'BeautifulSoup' se refiere a código HTML estructural o sintácticamente incorrecto escrito para una pagina web, lo uqe hace beautifulsoup es hacer la pagína hermosa y extrar información de ella con facilidad.


A continuación veremos breves ejemplos de como usarlo:

Una vez más, utilizaremos `requests` para raspar HTML de la web,luego creamos un objeto BeautifulSoup apartir del HTML resultante y lo guardamos en una variable.



```
from bs4 import BeautifulSoup
import requests
url = 'https://chilangdon.blogspot.com/2020/08/ciencia-de-datos-para-no-cientificos.html'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)
```

Al imprimir `soup.prettify` podemos ver que, por ejemplo la sopa pretificada esta sangrada en la forma correcta en la que esperariamos que se escribiera HTML correctamente.


```
print(soup.prettify())
```

A continuación exploraremos algunas funciones que pueden extraer el texto o el titulo.

```
print(soup.title)
```

```
print(soup.get_text())
```

También podermos usar `soup.find_all`, que extrar todas las URL de todos los hipervoinculos en HMTML.

```
for link in soup.find_all():
  print(link.get('href'))
```
