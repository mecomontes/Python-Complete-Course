# Solicitudes HTTP para imporratar archivos de la web.

Para importar archivos desde la web utilizamos la función `urlretrieve` de las solicitudes de urlib, analicemos esto un poco y en el rpoceso , comprendamos algunas cosas sobre cómo funciona Internet.

Url , significa localizador de recursos universal o unifrome y todo lo que realmente son , son referencias a recursos web.La gran mayoria de las URL son direcciones web.Pero pueden referirse a algunas otras cosas, como los protocolos de transeferencia de archivos y el acceso a las bases dedatos.

Actualmente nos centraremos en aquellas URL que son direcciones web o las ubicaciones de sitios web.

Dicha URL consta de 2 partes, un identificador de protocolo htpp o https y un nombre de recurso como github.com,la combianción de protocolo y nombre de recurso, que especifica de forma exclusiva la dirección web.

Para explicar las URL introdujimos el acronimo HTTP , que significa Protocolo de transeferncia de Hipertexto, por ejemplo Wikipedia prporciona una excelente descripcion de HTTP.

El rptocolo de transferencia de hipertexto es una aplicación de protocolo para sistemas de información distribuidos, colaborativos entre otros.

HTTP es la base de la comunicación de datos para la Web,tengamos en cuenta que HTTPS es una forma más segura de HTTP.

Cada vez que visitamos un sitio web , en realidad estamos enviando una sol.icitud HTTP a un servidor,esta solicitud se conoce como solicitud GET, que el es tipo mas común de solicitud HTTP.

Para hacer una consulta, usamos la función `urlretrieve ` , poero también guardamos los datos releevantes localmente, a continuación descubriremos como obtener los datos HTML de una página web.

HTML , significa lenguaje de marcado de hipertexto y es el lenguaje de marcado estandar para la web, para extraer el html de la pagína de inicio de weikipedia importamos las funciones necesarios, despues especificamos la URL en una variable.

Despues empaquetamos la solicitud GET utilizando la funcion request , enviamos la solicitud y capturamos la respuesta utilizando urloipen, por ultimo cerramos la consulta.

```
from urlib.request import urlopen,Request
url = 'https://en.wikipedia.org/wiki/Empirical_distribution_function'
request = Request(url)
response = urlopen(request)
html = response.read()
response.close()
```

Ahora haremoslos mismo, utilizando un paquete de solicitudes que proporciona una API maravillosa para realizar solicitudes `Requests` , este nos permite enviar solicitudes HTTP organicas alimentadas con pasto pero si la necesidad de mano de obra.

Primero importamos `requests`, despues especificamosla URL, empaquetamos la solicitud , enviamos la solicitud y capturamos el texto.




```
import requests
url = 'https://en.wikipedia.org/wiki/Empirical_distribution_function'
r = requests.get(url)
text = r.text
text
```

Realizar solicitudes HTTP en Python usando solicitudes

Ahora que ya sabemos cómo hacer solicitudes HTTP utilizando el paquete urllib, descubriremos cómo hacer lo mismo con la biblioteca de solicitudes de nivel superior. Una vez más, haremos ping a los servidores Github para su página "https://github.com/".



```
# Importar paquetes
import requests

# Especificar URL: url
url = "https://github.com/"

# Enviar consulta y guardar contendo: r
r = requests.get(url)

# Extraer contenido : text
text = r.text

# Imprmimri html
print(text)
```
