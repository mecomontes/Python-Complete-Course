# Introdución a los archivos planos.

En ete curso, nos centraremos en la ingesta de datos, un paso fundamental en cualquier proyecto de ciencia de datos, ya que no podemos analizar lo que no podemos ver.

También aprenderemos un poco, algunas técnicas para limpiar datos desordenados, para hacer esto usaremos la biblioteca de Python Pandas.

La parte central de pandas es el Data Frame, recodemos que un DataFrame es una estructura de datos bidimensional, lo cual significa que tiene columnas , generalmente etiquetadas con nombres de variables & filas que también tienen etiquetas  conocidas como indices.

El índice predeterminado es el número de fila, pero podemos epecificar una columna como indice y también podemos usar muchos tipos de datos.

Si bien podemos crear Data Frames generalmente  cargaremos datos existentes.Pandas maneja muchos formatos de datos, pero comencemos con uno de los más básicos el archivo plano.

## Flat File.

Estos archivos son simples, lo que los hace populares para almacenar y compratir datos, se pueden exportar desde sitemas de gestión de bases de datos y hojas de cálculo.

En un archivo plano ,los datos se almacenan como texto sin formato donde cada linea en el archivo representa una fila , con valores de columna separados por un carácter elegido llamado *delimitador*.

Por lo general el delimitador es una coma y estos arcivos se llaman *CSV* o comma-separated values , pero se pueden usar otros caracteres.Una sola función de pandas carga todos los archivos planos , sin importar el limitador con la función `read_csv()`.

### ¿Pero que pasas si ese archivo usa un delimitador diferente?

Pandas nos permite importar cualquier archivo plano con `pd.read_csv()` y pasandole el argumento `sep`.

Asignamos nuestra ruta a una ariable pero como argumento extra incluimos el argumento sep, indicando que esta separado por un tabulador que se representa con `"\t"`

```
house_file = pd.read_csv("/content/sample_data/california_housing_test.csv",sep="\t")
```
