# Grafica de Funciones.

Para calcular la derivada de una funci贸n f en un punto dado x, una soluci贸n con python es usar la funci贸n sympy llamada **derivate** & la libreria `Matplotlib` para graficar. 


Consideremos la siguiente funci贸n: $f(x)=sin(x^2) -cos(x)cos(x)$.

```python
#Importamos las librerias a utilizar
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
from scipy import misc
```

Para poder definir nuestras funciones matematicas es necesario crear una funcion con nuestros valores y otra para derivar,aqui dejo algunos parametros basicos de funciones:
 
 * `def`->Parametro para definir nuestra funcion.
 
 * `f` -> Nombre de la funcion.
 
 * `(x)` -> parametro a evaluar.
 
 * `return` -> Valor que queremos que nos regrese.

```python
#Definimos una funcion & definimos otra para calcular suderivada
def f(x): 
    return np.sin(x**2) - np.cos(x)*np.cos(x)
def d(x):
    return derivative(f,x)
```

## Mostrando los graficos.

En esta seccion graficaremos la funcion y su derivada primero las 2 graficas juntas y despues cada grafica por separado.


### Graficas Juntas:

```python
#Graficos juntos 

#Definimos los parametros del eje x
y = np.linspace(-4,4)

#Graficamos la funcion , y definimos el color y la etiqueta que va a aparecer en la leyenda
plt.plot(y,f(y),color="blue",label='$f(x)=sin(x^2) -cos(x)cos(x)$')
#Graficamos la derivada , y definimos el color y la etiqueta que va a aparecer en la leyenda
plt.plot(y,d(y),color="purple",label='$ d/dx=2xcos(x^2)+sin(2x)$')

#Etiquetamos los ejes X & Y
plt.xlabel("$X$")
plt.ylabel("$Y$")

#Ponemos un titulo de nuestra grafica
plt.title("Grafica de la funcion $f(x)=sin(x^2) -cos(x)cos(x)$ & su derivada")


#Mostramos las ecuaciones por color .
plt.legend()

#Mostramos la grafica.
plt.show()
```

## Graficamos las graficas por separado.

Es importante mencionar que primero tenemos que definir el numero de graficas dado por filas y columnas , en este caso como son 2 graficas nuestros parametros seran:

 * `nrows=1`-> Ya que solo queremos 2 filas.
 * `ncols=2`-> Ya que queremos 1 columna.
 * `fig`  ->Este parametro es util si deseamos cambiar los atributos de nivel de nuestra figur`.
 
 * `plt.subplots()` -> Funcion que devuelve una tupla que contien una figura y ejes
 
 * `ax` -> Ejes.

```python
#Grafiacamos nuestras funciones:

#Definimos una figura y los ejes
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1)


#Definimos los parametros del eje x
y = np.linspace(-4,4)


#Graficamos la funcion a derivar 
ax1.plot(y,f(y),color="red",label='$f(x)=sin(x^2) -cos(x)cos(x)$')
ax1.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
ax1.set_ylabel("$Y$")
ax1.legend()


#Graficamos la derivada
ax2.plot(y,d(y),color="purple",label='$ d/dx=2xcos(x^2)+sin(2x)$')
ax2.grid(color="black", alpha=0.5, linestyle='dashed', linewidth=0.5)
ax2.set_ylabel("$Y$")
ax2.legend()

plt.show()
```
