Es f√°cil hojear los dos conjuntos de tendencias y detectar tendencias comunes, pero no hagamos un trabajo "manual".
Podemos usar alguna estructura de datos de conjuntos de Python para encontrar tendencias comunes; podemos iterar a trav√©s de los dos objetos de tendencias, convertir las listas de nombres en conjuntos y llamar al m√©todo de intersecci√≥n para obtener los nombres comunes entre los dos conjuntos.

### INSTRUCCIONES:

* Extraer el campo de nombre, `trend['name']`, de la lista de tendencias en `WW_trends` y `US_trends` usando la lista de comprensi√≥n (*). Puedes usar `WW_trends[0]['tendencias']` y `US_trends[0]['tendencias']` para que las iteraciones obtengan los nombres porque los objetos de tendencias son listas con un solo elemento.

* Llamar al m√©todo `intersection()` en `world_trends` con `us_trends` como par√°metro de entrada para obtener los elementos comunes entre los dos; almacenar la salida en la variable llamada `common_trends``.

```python
# Extraer todos los nombres de tendencias WW de WW_trends
world_trends = set([trend['name']
                    for trend in WW_trends[0]['trends']])

# Extraer todos los nombres de tendencias de EE. UU. De US_trends
us_trends = set([trend['name']
                 for trend in US_trends[0]['trends']]) 

# Obtener la intersecci√≥n de los dos conjuntos de tendencias
common_trends = world_trends.intersection(us_trends)

# Inspeccionanado nuestros datos
print(world_trends,"\n")
```

# Cavando m√°s profundoüïµÔ∏è‚Äç‚ôÄÔ∏è

La impresi√≥n de los dos primeros elementos del tweet nos hace darnos cuenta de que hay mucho m√°s en un tweet de lo que normalmente consideramos un tweet: ¬°hay mucho m√°s que un texto breve!

Pero bueno, ¬°no nos dejemos abrumar por toda la informaci√≥n en un objeto de tweet! Centr√©monos en algunos campos interesantes y veamos si podemos encontrar informaci√≥n oculta all√≠.

### INSTRUCCIONES:

* Para cada tweet en el objeto `tweets`, extraer su campo de texto, `tweet['text']`, usando la comprensi√≥n de lista. Almacenar todos los textos de salida en una lista denominada `texts`.


* Para cada tweet en tweets, crear un bucle interno para iterar a trav√©s de las menciones de usuario, `tweet ['entities']['user_mentions']`. De cada menci√≥n_usuario extraiga su campo de nombre de pantalla, user_mention['screen_name']. Almacenar la salida en `names`.


* Para cada tweet en tweets, cree un bucle interno para iterar a trav√©s de hashtags, tweet ['entidades'] ['hashtags']. De cada hashtag extraiga su campo de texto, hashtag ['texto']. Almacena la salida en hashtags

```python
# Extraer el texto de todos los tweets del objeto tweet
texts = [tweet['text'] 
                 for tweet in tweets ]

# Extraer los nombres de pantalla de los usuarios que tuitean sobre #WeLoveTheEarth
names = [user_mention['screen_name'] 
                 for tweet in tweets
                     for user_mention in tweet['entities']['user_mentions']]

# Extraer todos los hashtags que se utilizan al hablar de este tema.
hashtags = [hashtag['text'] 
             for tweet in tweets
                 for hashtag in tweet['entities']['hashtags']]
```

# An√°lisis de frecuenciaüïµÔ∏è‚Äç‚ôÄÔ∏è 

Solo de los primeros resultados de la √∫ltima extracci√≥n, podemos deducir que:

Ahora podemos echar un vistazo m√°s de cerca haciendo un ejercicio simple, pero muy √∫til: calcular las distribuciones de frecuencia. Comenzar de manera simple con frecuencias es generalmente un buen enfoque; ayuda a obtener ideas sobre c√≥mo seguir adelante.

### INSTRUCCIONES:

* Importe el m√≥dulo Counter de colecciones. 

* Llamar al m√©todo Counter () con el elemento del bucle for como par√°metro de entrada. (Esto le permite realizar un seguimiento de cu√°ntas veces se agregan los mismos valores).

```python
# importando counter
from collections import Counter

# Contando ocurrencias / obteniendo la distribuci√≥n de frecuencia de todos los nombres y hashtags
for item in [names, hashtags]:
    c = Counter(item)    
    # Inspeccionando los 10 elementos m√°s comunes en c
    print (c.most_common(10), "\n")
```
